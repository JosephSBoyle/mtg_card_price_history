```python
price_dicts: list[dict[str, float]] = await get_price_dicts()
```

```python
one_card = price_dicts[50]
datetimes = [dt.datetime.strptime(d, "%Y-%m-%d").date() for d in one_card.keys()]

plt.gca().xaxis.set_major_locator(mdates.WeekdayLocator()) # reduce the frequency of x-axis ticks
plt.gcf().autofmt_xdate()  # format / rotate the dates
plt.plot(one_card.keys(), one_card.values())
```
